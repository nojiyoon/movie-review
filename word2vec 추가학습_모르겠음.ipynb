```python
import pickle
import gensim
from gensim.models import Word2Vec, KeyedVectors


model = gensim.models.Word2Vec.load("./model/ko.bin")
model.wv.save_word2vec_format("ko.bin.gz", binary=False)

# model2 = Word2Vec.load('./model/review_word2vec.model')

with open('./model/tokenizer_test.pickle', 'rb') as f:
    tokenizer = pickle.load(f)
tokenized_data = list(tokenizer.word_index)
    
model_2 = Word2Vec(size=200, min_count=1)
model_2.build_vocab(tokenized_data)
total_examples = model_2.corpus_count
print(total_examples)
```

```python
model_2.build_vocab([list(model.wv.vocab.keys())], update=True)
model_2.intersect_word2vec_format("ko.bin.gz", binary=False)
```

```python
model_2.train(tokenized_data, total_examples=total_examples, epochs=model_2.iter)
```

```python
result = model_2.wv.most_similar('돈')
print(result)
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
text = 'i love music.'

tokenizer = RegexpTokenizer(r'\w+')    
tokenized_text = [tokenizer.tokenize(i) for i in text]
```

```python
tokenized_text
```

```python
model_2 = Word2Vec(size=300, min_count=1)
 
model_2.build_vocab(tokenized_text)
 
total_examples = model_2.corpus_count
```
